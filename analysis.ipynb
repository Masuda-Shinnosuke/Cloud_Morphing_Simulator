```python
import json
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('/home/masuda/Doxuments/CMorph/output/serverdata.json') as json_file:
    data = json.load(json_file)
```

```python
server0=[]
server1=[]
server2=[]
server3=[]
for i in range(0,len(data),4):
    step_data=data[i:i+4]
    server0.append(step_data[0]["currentRho"])
    server1.append(step_data[1]["currentRho"])
    server2.append(step_data[2]["currentRho"])
    server3.append(step_data[3]["currentRho"])
```

```python
# plt.plot(server0[100:200],label="Server0")
# plt.plot(server1[100:200],label="server1")
# plt.plot(server2[100:200], label="Server2")
# plt.plot(server3[100:200],label="server3")

plt.plot(server0,label="Server0")
plt.plot(server1,label="server1")
plt.plot(server2, label="Server2")
plt.plot(server3,label="server3",)

# 
plt.xlabel("Step")
plt.ylabel("Current Rho")
plt.title("Server Data over Steps")
plt.legend()  # 

# 
plt.show()
# 
```
